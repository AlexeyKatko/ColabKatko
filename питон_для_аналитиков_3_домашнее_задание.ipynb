### –ó–∞–¥–∞–Ω–∏–µ 1.2

–°–æ–∑–¥–∞—Ç—å –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫ delta_renovated, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ä–∞–∑–Ω–∏—Ü—É –≤ –≥–æ–¥–∞—Ö –º–µ–∂–¥—É –≥–æ–¥–æ–º —Ä–µ–Ω–æ–≤–∞—Ü–∏–µ–π –¥–æ–º–∞ –∏ –≥–æ–¥–æ–º –ø–æ—Å—Ç—Ä–æ–π–∫–∏ –¥–æ–º–∞

–ï—Å–ª–∏ —Ä–µ–Ω–æ–≤–∞—Ü–∏–∏ –¥–æ–º–∞ –Ω–µ –±—ã–ª–æ, —Ç–æ –≤ –Ω–æ–≤–æ–º –ø—Ä–∏–∑–Ω–∞–∫–µ –ø–æ—Å—Ç–∞–≤—Ç–µ 0

```python
yr_diff = []

for i, row in df.iterrows():
    if row['yr_renovated']:
        yr_diff.append(row['yr_renovated']-row['yr_built'])
    else:
        yr_diff.append(0)

len(yr_diff)
```

```python
df['delta_renovated'] = yr_diff

df[['yr_built', 'yr_renovated', 'delta_renovated']].head()
```

```python
df[df['yr_renovated'] > 0][['yr_built', 'yr_renovated', 'delta_renovated']].head()
```
